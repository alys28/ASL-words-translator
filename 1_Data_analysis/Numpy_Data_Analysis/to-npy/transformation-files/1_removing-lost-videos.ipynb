checking if video is unavailable, using requests


```python
import requests

def check_video_url(video_id):
    checker_url = "https://www.youtube.com/oembed?url=http://www.youtube.com/watch?v="
    video_url = checker_url + video_id

    request = requests.get(video_url)

    return request.status_code == 200

test="NjM7VsurXto"
print(check_video_url(test))

test="-DMFu-nAsok"
print(check_video_url(test))
```

Removing unavailable videos


```python
test_df.head()

#boolean list containing if a certain values if availble.
    #all videos will be marked as not available for the moment.
test_df["available"]=False
itr = next(test_df.iterrows())[1]
test_df["url"].head()
test_df["available"].head()
test_df.head()


```

```python
#if available, make the last column true;

for idx in range(0, len(test_df)-1):
    #last eleven elements
    code=test_df.loc[idx, "url"][-11:]#[24:]
    # print(code)
    if idx%1000==0:
        print(idx)
    if(check_video_url(code)):
        test_df.loc[idx, "available"]=True
```
